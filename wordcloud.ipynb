```python
file = open("cloud1.txt", 'r')
lists = file.readlines()
file.close()
lists

type(lists)
```

```python
from collections import Counter
count = Counter(lists)
```

```python
words = dict(count.most_common())
```

```python
from wordcloud import WordCloud 

import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
%matplotlib inline

import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')

set_matplotlib_formats('retina')

matplotlib.rc('axes',unicode_minus = False)
```

```python
group_ing = []  # 모든 원료
for i in lists:
    item = i.split(',')  # ,로 여러 원료 구분x
    for j in item:
        group_ing.append(j.strip())  # 전체리스트에 추가
        
print(group_ing)
```

```python
dic = {}
for w in group_ing:
    if w in dic:
        dic[w] += 1
    else:
        dic[w] = 1

lst = sorted(dic.items(), key=lambda kv: kv[1], reverse=True)
for w, c in lst[:31]:
    print(w, c)
```

```python
wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf',background_color='white',colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(dic)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
